```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from yellowbrick.cluster import KElbowVisualizer
```

```python
#Label Encoding the object dtypes.
LE=LabelEncoder()
for i in object_cols:
    data2[i]=data2[[i]].apply(LE.fit_transform)

print("All features are now numerical")
```

```python
#Creating a copy of data
ds = data2.copy()
# creating a subset of dataframe by dropping the features on deals accepted and promotions
cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
ds = ds.drop(cols_del, axis=1)

# Check for datetime columns and convert them to numerical
for col in ds.columns:
    if ds[col].dtype == 'datetime64[ns]':
        ds[col] = ds[col].astype('int64')

#Scaling
scaler = StandardScaler()
scaler.fit(ds)
scaled_ds = pd.DataFrame(scaler.transform(ds),columns= ds.columns )
print("All features are now scaled")
```

```python
#Scaled data to be used for reducing the dimensionality
print("Dataframe to be used for further modelling:")
scaled_ds.head()
```

```python
#Dropping ID as it doesnt contribute to any significance
scaled_ds = scaled_ds.drop("ID", axis=1)

scaled_ds
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances
```

```python
#3 clusters
pca = PCA(n_components=3)
pca.fit(scaled_ds)
PCA_ds = pd.DataFrame(pca.transform(scaled_ds), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```
